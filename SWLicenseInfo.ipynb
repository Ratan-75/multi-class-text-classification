```python
import json
```

```python
from difflib import get_close_matches
```

```python
data = json.load(open("SWLicenseInfo.json"))
```

```python
data
```

```python
def getLicenseInfo(swname):
    if swname in data:
        return data[swname]
    elif swname.title() in data:
        return data[swname.title()]
    elif swname.upper() in data:
        return data[swname.upper()]
    elif swname.lower() in data:
        return data[swname.lower()]
    elif len(get_close_matches(swname, data.keys())) > 0 :
        ask = input("Did you mean %s instead? Enter Y if yes, or N if no: "
                    %(get_close_matches(swname, data.keys())[0]))
        if ask == "Y":
            return data[get_close_matches(swname, data.keys())[0]]
        elif ask == "N":
            return "This software product does not exist. Please double check"
        else:
            return "Apologies, We didn't understand the input provided."
    else:
        return "This software product does not exist. Please double check"
```

```python
productName = input("Enter the product name: ") 
output = getLicenseInfo(productName)
if type(output) == list:
    for item in output:
        print(item)
else:
    print(output)
```
